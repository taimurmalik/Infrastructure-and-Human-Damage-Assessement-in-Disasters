```python
# âœ… Drive Mount
from google.colab import drive
drive.mount('/content/drive')

# âœ… GPU Info Logging
import tensorflow as tf
import os
!nvidia-smi
print("\nâœ… TensorFlow version:", tf.__version__)
print("âœ… GPU device name:", tf.config.list_physical_devices('GPU'))

# âœ… Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import glob
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers.schedules import CosineDecayRestarts
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# âœ… Setup GPU
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    print("âœ… Using GPU")
else:
    print("âš  Using CPU")

# âœ… Data Paths
base_path = "/content/drive/MyDrive/fyp_checkpoints"
os.makedirs(base_path, exist_ok=True)
train_dir = "/content/drive/MyDrive/split_data/train"
val_dir = "/content/drive/MyDrive/split_data/val"
test_dir = "/content/drive/MyDrive/split_data/test"

# âœ… Remove hidden files
!find /content/drive/MyDrive/split_data/ -name '.*' -delete

# âœ… Inspect directory structure
print("\nInspecting directory structure:")
base_dirs = [train_dir, val_dir, test_dir]
for base_dir in base_dirs:
    print(f"\nChecking {base_dir}:")
    if os.path.exists(base_dir):
        subdirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
        print(f"Subdirectories found: {subdirs}")
        for subdir in subdirs:
            folder_path = os.path.join(base_dir, subdir)
            files = glob.glob(os.path.join(folder_path, '*.jpg')) + \
                    glob.glob(os.path.join(folder_path, '*.jpeg')) + \
                    glob.glob(os.path.join(folder_path, '*.png'))
            print(f"  {subdir}: {len(files)} files found (e.g., {files[:2]})")
    else:
        print(f"{base_dir} does not exist")

# âœ… Define the mapping from disaster events to the 3 labels present in the data
human_mapping = {
    'California_Wildfires': 3,  # "non-humanitarian"
    'Hurricane_Harvey': 2,     # "rescue and volunteering"
    'mexico_earthquake': 0,    # "1 injured or dead"
    'iraq_iran_earthquake': 0, # "1 injured or dead"
    'srilanka_floods': 3,      # "non-humanitarian"
    'Hurricane_Maria': 2,      # "rescue and volunteering"
    'Hurricane_Irma': 2        # "rescue and volunteering"
}

# âœ… Utility function: Create DataFrame with enhanced debugging
def create_dataframe(data_dir, mapping):
    filepaths = []
    labels = []

    # Check if the directory exists
    if not os.path.exists(data_dir):
        raise FileNotFoundError(f"Directory {data_dir} does not exist. Please check the path.")

    # Check each class folder
    for class_name, label in mapping.items():
        folder_path = os.path.join(data_dir, class_name)
        if not os.path.exists(folder_path):
            print(f"Warning: Folder {folder_path} does not exist. Skipping...")
            continue

        files = glob.glob(os.path.join(folder_path, '*.jpg')) + \
                glob.glob(os.path.join(folder_path, '*.jpeg')) + \
                glob.glob(os.path.join(folder_path, '*.png'))

        if not files:
            print(f"Warning: No image files found in {folder_path}. Skipping...")
            continue

        print(f"Found {len(files)} files in {folder_path}")
        filepaths.extend(files)
        labels.extend([str(label)] * len(files))

    # Create DataFrame
    df = pd.DataFrame({'filename': filepaths, 'class': labels})

    # Check if DataFrame is empty
    if df.empty:
        raise ValueError(f"No data found in {data_dir}. Please check the directory structure and files.")

    return df

# âœ… Create DataFrames with error handling
try:
    train_df = create_dataframe(train_dir, human_mapping)
    val_df = create_dataframe(val_dir, human_mapping)
    test_df = create_dataframe(test_dir, human_mapping)
except Exception as e:
    print(f"Error creating DataFrames: {e}")
    raise

# Debugging: Check the DataFrame structure
print("\nTraining DataFrame:")
print(train_df.head())
print("\nTraining DataFrame columns:", train_df.columns.tolist())
print("\nTraining DataFrame distribution:")
print(train_df['class'].value_counts())

# âœ… Data Generators with enhanced augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.25,
    height_shift_range=0.25,
    zoom_range=[0.8, 1.2],
    shear_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Setup generators with error handling
try:
    train_generator = train_datagen.flow_from_dataframe(
        train_df,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        shuffle=True
    )
    val_generator = val_test_datagen.flow_from_dataframe(
        val_df,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
    test_generator = val_test_datagen.flow_from_dataframe(
        test_df,
        x_col='filename',
        y_col='class',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
except Exception as e:
    print(f"Error setting up data generators: {e}")
    raise

# âœ… Compute Class Weights
try:
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(train_df['class']),
        y=train_df['class']
    )
    class_weights = {i: w for i, w in enumerate(class_weights)}
    print("Computed class weights:", class_weights)
except Exception as e:
    print(f"Error computing class weights: {e}")
    raise

# âœ… Build VGG16-based model with fine-tuning
base_model = VGG16(include_top=False, weights='imagenet', input_shape=(224, 224, 3))

# Unfreeze the last few layers of VGG16 for fine-tuning
for layer in base_model.layers[:-4]:
    layer.trainable = False
for layer in base_model.layers[-4:]:
    layer.trainable = True

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.4)(x)
output = Dense(3, activation='softmax')(x)  # Only 3 labels present in the data
model = Model(inputs=base_model.input, outputs=output)

# âœ… Model Summary
model.summary()

# âœ… Compile with optimized learning rate schedule
lr_schedule = CosineDecayRestarts(
    initial_learning_rate=1e-4,
    first_decay_steps=500,
    t_mul=2.0,
    m_mul=0.9,
    alpha=1e-5
)
optimizer = Adam(learning_rate=lr_schedule)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# âœ… Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_ckpt_best = ModelCheckpoint(
    os.path.join(base_path, "best_model.h5"),
    save_best_only=True,
    monitor='val_accuracy',
    mode='max'
)

# âœ… Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=50,
    class_weight=class_weights,
    callbacks=[early_stop, model_ckpt_best],
    verbose=1
)

# âœ… Save Final Model
final_model_path = os.path.join(base_path, "final_trained_model.h5")
model.save(final_model_path)
print(f"âœ… Final model saved to: {final_model_path}")

# âœ… Convert to TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
with open(os.path.join(base_path, "model_flutter.tflite"), "wb") as f:
    f.write(tflite_model)
print("âœ… TFLite model saved")

# âœ… Evaluation on Validation Set
y_true_val = val_generator.classes
y_pred_val = model.predict(val_generator)
y_pred_val_classes = np.argmax(y_pred_val, axis=1)

val_acc = accuracy_score(y_true_val, y_pred_val_classes)
val_prec = precision_score(y_true_val, y_pred_val_classes, average='weighted')
val_rec = recall_score(y_true_val, y_pred_val_classes, average='weighted')
val_f1 = f1_score(y_true_val, y_pred_val_classes, average='weighted')

print("\nğŸ“Š Validation Set Evaluation for Human Class:")
print(f"Accuracy: {val_acc:.4f} | Precision: {val_prec:.4f} | Recall: {val_rec:.4f} | F1 Score: {val_f1:.4f}")
print("\nğŸ§¾ Classification Report (Validation) for Human Class:")
print(classification_report(y_true_val, y_pred_val_classes, target_names=["1 injured or dead", "rescue and volunteering", "non-humanitarian"]))

# âœ… Evaluation on Test Set
model = load_model(os.path.join(base_path, "best_model.h5"))
y_true_test = test_generator.classes
y_pred_test = model.predict(test_generator)
y_pred_test_classes = np.argmax(y_pred_test, axis=1)

cm = confusion_matrix(y_true_test, y_pred_test_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["1 injured or dead", "rescue and volunteering", "non-humanitarian"], yticklabels=["1 injured or dead", "rescue and volunteering", "non-humanitarian"])
plt.title('Confusion Matrix (Test Set) for Human Class')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()

test_acc = accuracy_score(y_true_test, y_pred_test_classes)
test_prec = precision_score(y_true_test, y_pred_test_classes, average='weighted')
test_rec = recall_score(y_true_test, y_pred_test_classes, average='weighted')
test_f1 = f1_score(y_true_test, y_pred_test_classes, average='weighted')

print("\nâœ… Test Set Evaluation for Human Class:")
print(f"Accuracy: {test_acc:.4f} | Precision: {test_prec:.4f} | Recall: {test_rec:.4f} | F1 Score: {test_f1:.4f}")
print("\nğŸ§¾ Classification Report (Test) for Human Class:")
print(classification_report(y_true_test, y_pred_test_classes, target_names=["1 injured or dead", "rescue and volunteering", "non-humanitarian"]))

# âœ… Visualize Metrics
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
